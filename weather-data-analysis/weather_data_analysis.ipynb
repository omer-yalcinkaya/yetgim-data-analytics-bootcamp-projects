**1. Kütüphane ve Veri Yükleme:**

Pandas ve Numpy kütüphanelerini import edilmesi.

```python
import numpy as np
import pandas as pd
```

weather_data.csv dosyasındaki verilerin yüklenmesi ve ekranda gösterilmesi.

```python
df= pd.read_csv('weather_data.csv')
print(df)
```

**2. Veriyi Keşfetme:**

Veri setinin ilk 5 satırının(head) ve son 5 satırının(tail) listelenmesi.

```python
combined=pd.concat([df.head(5),df.tail(5)] )
print(combined)
```

Veri setinin istatistiksel özetinin(describe) görüntülenmesi.

```python
df.describe()
```

**3. Sütün Seçimi**

Sadece Date, City ve Temperature sütunlarını seçerek yeni bir görünüm oluşturup ve ekrana yazdırılması.

```python
df[['Date','City','Temperature']]
```

City ve Temperature kolonlarını birlikte listelenmesi.


```python
df[['City', 'Temperature']]
```

**4. Basit Filtreleme**



Hava sıcaklığı (Temperature) 30 derecenin üzerinde olan kayıtların listelenmesi.

```python
df[df['Temperature']>30]
```

Şehri (City) sadece "Bursa" olan kayıtları listelenmesi.

```python
df[df['City']=='Bursa']
```

**5. Mantıksal Operatörler ile Filtreleme:**



Şehri "İstanbul" olan VE Nem oranı (Humidity) 60'tan büyük olan kayıtların listelenmesi.

```python
df[(df['City']=='İstanbul') & (df['Humidity']>60)]
```

 Şehri "Ankara" olan veya Sıcaklığı 5 dereceden küçük olan kayıtların listelenmesi.

```python
df[(df['City']=='Ankara') | (df['Temperature']<5)]
```

Sıcaklığı 10°C altında veya nemi (Humidity) %70 üzerinde olan verileri filtreleyiniz.

```python
df[(df['Temperature']<10) | (df['Humidity'] >70 )]
```

**6. Sıralama (Sorting):**

Veri setini Temperature (Sıcaklık) değerine göre büyükten küçüğe (azalan) sıralanıp ilk 10 kaydının listelenmesi.

```python
df.sort_values('Temperature',ascending=False).head(10)
```

Veriyi neme(Humidity) göre azalan şekilde sıralanması.

```python
df.sort_values('Humidity',ascending=False)
```

Veriyi Şehir alanına göre Artan şekilde sıralanması.

```python
df.sort_values('City')
```

**7- Yeni Sütun Ekleme: **

Veri setine Temperature_F adında yeni bir sütununun eklenip. Bu sütun, sıcaklığın Fahrenheit cinsinden değerini taşıması. Formül: Temperature_F = (Temperature * 9/5) + 32

```python
df['Temperature_F']=(df['Temperature']* 9 / 5)+32
df
```

"FeelsLike" isminde yeni bir kolon oluşturun. Formül: FeelsLike = Temperature - (Humidity / 100)

```python
df['FeelsLike']=df['Temperature']-(df['Humidity']/100)
df
```

**8. Gruplama(Grouping) ve Analiz:**





Her şehirde (City) kaç adet veri kaydı olduğunun  hesaplanması.

```python
df.groupby('City')['Date'].count()
```

Şehirlere göre ortalama sıcaklık (Temperature) değerlerinin hesaplanması.

```python
avg_temp_by_city=df.groupby('City')['Temperature'].mean()
avg_temp_by_city
```

**9. En Yüksek/Düşük Değer Analizi:**

 En yüksek sıcaklığın olduğu satırı (tüm bilgilerini) bulup yazdırılması.

```python
df.loc[df['Temperature'].idxmax()]
```

En düşük nem oranının olduğu satırı bulunup yazdırılması.

```python
df.loc[df['Humidity'].idxmin()]
```

8. maddede bulduğunuz "Şehirlere göre ortalama sıcaklık" tablosunu bir CSV dosyası olarak kaydedilmesi.
